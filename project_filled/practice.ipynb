```python
import os, shutil, random
```

```python
dataset_root = '/Users/arushreemishra/Downloads/brain_tumor/project_filled/brain_tumor_dataset'
working_root = '/Users/arushreemishra/Downloads/brain_tumor/project_filled/dataset'
train_root = os.path.join(working_root, 'train')
val_root = os.path.join(working_root, 'val')
test_root = os.path.join(working_root, 'test')  

os.makedirs(train_root, exist_ok=True)
os.makedirs(val_root, exist_ok=True)
os.makedirs(test_root, exist_ok=True)
```

```python
val_ratio = 0.2  # 20% for validation
classes = [d for d in os.listdir(os.path.join(dataset_root, 'Training')) if os.path.isdir(os.path.join(dataset_root, 'Training', d))]

for cls in classes:
    cls_path = os.path.join(dataset_root, 'Training', cls)
    files = [f for f in os.listdir(cls_path) if f.endswith(('.jpg','.png','.jpeg'))]
    random.shuffle(files)
    split_idx = int(len(files) * val_ratio)
    
    os.makedirs(os.path.join(train_root, cls), exist_ok=True)
    os.makedirs(os.path.join(val_root, cls), exist_ok=True)
    
    for i, f in enumerate(files):
        src = os.path.join(cls_path, f)
        if i < split_idx:
            dst = os.path.join(val_root, cls, f)
        else:
            dst = os.path.join(train_root, cls, f)
        shutil.copy(src, dst)
```

```python
test_classes = [d for d in os.listdir(os.path.join(dataset_root, 'Testing')) if os.path.isdir(os.path.join(dataset_root, 'Testing', d))]
for cls in test_classes:
    cls_path = os.path.join(dataset_root, 'Testing', cls)
    os.makedirs(os.path.join(test_root, cls), exist_ok=True)
    for f in os.listdir(cls_path):
        shutil.copy(os.path.join(cls_path, f), os.path.join(test_root, cls, f))
```

```python
print("Train classes:", os.listdir(train_root))
print("Val classes:", os.listdir(val_root))
print("Test classes:", os.listdir(test_root))
```
